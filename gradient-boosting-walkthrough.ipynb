```python
from sklearn.datasets import load_wine
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import polars as pl
```

### Preparing the data

```python
# Load the wine dataset from the sklearn library
dataset = load_wine()

# Returns a dict-like object
print(dataset.keys())
```

```python
# Returns a numpy array of shape (training_examples, features)
X = dataset['data']
print(f"X: {X.shape}")
```

```python
# Returns an array of shape (training_examples)
y = dataset['target']
print(f"y: {y.shape}")
```

```python
# Randomly shuffles and splits the data into 80% training / 20% testing
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size=0.20, random_state=0)
print(f"Training examples: {X_train.shape[0]}, Testing_examples: {X_test.shape[0]}")
```

### Training the model

```python
model = GradientBoostingClassifier(
    n_estimators = 100,
    learning_rate = 0.1,
    random_state = 0,
)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
